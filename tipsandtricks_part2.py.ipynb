# 9 - Convert Strings to number 

```python
df=pd.DataFrame({'Col_A':[1,2,3,6,7,8,9,22],'Col_B':[4,5,6,7,8,9,10,11]})
df
```

```python
df.dtypes
```

```python
df=pd.DataFrame({'Col_A':['1','2','3','6','7','8','9','22'],'Col_B':['4','5','6','7','8','9','10','11']})
df
```

```python
df.dtypes
```

```python
df.astype({'Col_A':'int64','Col_B':'int64'}).dtypes
```

```python
df.astype({'Col_A':'float64','Col_B':'int64'}).dtypes
```

```python
pd.to_numeric(df['Col_A'],errors='coerce')
```
